rubric={reasoning:60}

<div class="alert alert-info">

<li>What is representation bias?</li>
<li>What is measurement bias?</li>
<li>What is aggregation bias?</li>
<li>What is evaluation bias?</li>
<li>What is deployment bias?</li>

YOUR ANSWERS HERE

1. Historical bias occurs when the data used to train an AI system reflects existing prejudices, inequalities, or stereotypes that are already present in society and the real world. Even if the data is perfectly sampled and measured, it can encode longstanding societal inequities like for example, a hiring model trained on a workforce that historically favored men will learn to prefer male candidates.

2. Representation bias arises when the population used to create a training dataset does not accurately reflect the broader population the model will be applied to eventually. For example, if a facial recognition system is trained primarily on lighter skinned faces, it will perform poorly on darker skinned individuals because they were under represented in the data, as simple as that. 

3. Measurement bias happens when the features or labels chosen as proxies for the variables of interest are inaccurate or inconsistent across different groups. For example, using arrest rates as a proxy for crime rates in predictive policing can unfairly target over policed communities, because arrest data reflects policing patterns rather than actual criminal behavior.

4. Aggregation bias occurs when a single model is applied to a diverse population that actually contains distinct subgroups with different characteristics. By treating everyone the same, the model can perform poorly for specific subgroups, such as using a single diabetes prediction threshold across different ethnicities when the relevant biomarker levels vary by ethnicity.

5. Evaluation bias arises when the benchmark or test data used to evaluate a model does not properly represent the full population or the actual use case. If a model is assessed using a dataset that over represents certain demographics, it may appear accurate overall while underperforming for under represented groups.

6. Deployment bias occurs when a model is used in a context or for a purpose that differs from what it was originally designed and evaluated for. For example, a model built to assist human decision makers may instead be used as the sole decision maker, leading to unintended and potentially harmful outcomes.

YOUR ANSWERS HERE

1. There are actually some ways that algorithmic bias can be more useful than human bias when it comes to figuring out where the problem is and fixing it. Algorithms are auditable, you can look at the code and the decision logs, whereas human biases are often unconscious and the person doesnt even realize they have them. Algorithmic decisions are also reproducible, same input gives same output, which means you can run controlled tests to pinpoint exactly where the bias is coming from. And once you find and fix a bias in an algorithm, that fix applies to every future decision at scale, retraining hundreds of human employees is way slower and harder to verify. So in a weird way, a biased algorithm can actually serve as a diagnostic tool that reveals systemic discrimination patterns that were hidden when individual humans were making the calls. 

2. For auditing this rare disease detection algorithm, we'd consider all six biases from question 1.1.

- Historical bias is relevant because the medical data used probably reflects gaps in healthcare access. Some populations might have been historically underdiagnosed simply because they didnt have access to specialists.

- Representation bias is probably the most critical one here. we'd want to know exactly who was in the training data, if it came from one region or mostly one ethnic group, the model could fail on a national population, especially since its a rare disease. 

- Measurement bias matters because skeletal structure varies naturally across ethnicities, ages and body types. The algorithm might mistake normal anatomical variation in one group for a defect, and X-ray quality also varies a lot between hospitals which could affect performance.

- Aggregation bias is worth looking at because a single model for the entire national population might not account for how the disease manifests differently across demographic groups. I'd ask the company whether they tested performance separately for different subgroups.

- Evaluation bias is a big one here. The 99.9% accuracy claim is suspicious for a rare disease, it could just mean the model predicts "no disease" almost every time and gets it right because the disease is so uncommon.

- Deployment bias is also important since the algorithm was developed in a controlled setting but would be used across the whole national healthcare system. Different hospitals have different equipment and patient populations, and clinics might treat the model's output as a definitive diagnosis rather than a screening tool.

YOUR ANSWERS HERE

1. Whether this tradeoff is worth it really depends on the context. In high stakes domains like healthcare or even criminal justice where decisions directly affect peoples lives, accepting a small drop in overall accuracy to avoid systematically harming a specific group is pretty much always justified. Its also worth considering that high overall accuracy can be misleading, it often just masks terrible performance on minority subgroups, so balancing the data actually gives you a more honest picture of how the model really performs. The magnitude matters too, if subgroup performance improves a lot and overall performance only dips slightly, thats an easy call. And you should think about who bears the cost versus who benefits, if the cost is spread broadly but the benefit goes to a group that already faces systemic disadvantage, that makes it more worthwhile in doing so. I think a good general rule is that the tradeoff is most justified when the harms of biased predictions are severe and hard to reverse, and least justified when the application is low stakes and the performance gap would be large.
